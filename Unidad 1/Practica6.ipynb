```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
from google.colab import files
```

```python
df = pd.read_csv("autos2.csv")
```

```python
print("Primer fila de la base de datos:")
print(df.head())
```

```python
print("\nEstadsticas bsicas:")
print(df.describe())
```

```python
precioMin = df['price'].min()
precioMax = df['price'].max()
precioModa = df['price'].mode()[0]
precioMedia = df['price'].mean()
precioMediana = df['price'].median()
precio_std = df['price'].std()
```

```python
print("\nMedidas de tendencia central de precios:")
print(f"Mnimo: {precioMin}")
print(f"Mximo: {precioMax}")
print(f"Moda: {precioModa}")
print(f"Media: {precioMedia}")
print(f"Mediana: {precioMediana}")
print(f"Desviacin estndar: {precio_std}")
```

```python
num_bins = math.ceil(1 + 3.322 * math.log10(len(df)))
plt.figure(figsize=(10, 5))
plt.hist(df['price'], bins=num_bins, edgecolor='black', alpha=0.7)
plt.xlabel('Precio del automvil')
plt.ylabel('Frecuencia')
plt.title('Histograma de los precios de automviles')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='body-style', y='price', data=df)
plt.xlabel('Tipo de vehculo')
plt.ylabel('Precio del automvil')
plt.title('Diagrama de cajas y bigotes: Precio vs Tipo de Vehculo')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='drive-wheels', y='price', data=df)
plt.xlabel('Traccin del vehculo')
plt.ylabel('Precio del automvil')
plt.title('Diagrama de cajas y bigotes: Precio vs Traccin del Vehculo')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='engine-location', y='price', data=df)
plt.xlabel('Ubicacin del motor')
plt.ylabel('Precio del automvil')
plt.title('Diagrama de cajas y bigotes: Precio vs Ubicacin del Motor')
plt.show()
```
